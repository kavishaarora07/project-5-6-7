# now the data is balanced through smote technique

since the outcome has two variables it is binary model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
lr=LogisticRegression()
```

# splitting the dataframe

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print("at random state",i,"the model performs v well")
        print("at random state:-",i)
        print("training accuracy is",accuracy_score(y_train,pred_train)*100)
        print("testing accuracy",accuracy_score(y_test,pred_test)*100)
        
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=882)
```

```python
import sklearn
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test))
```

# cross validation for logistic regression

```python
pred_lr=lr.predict(x_test)
from sklearn.model_selection import cross_val_score
lss=accuracy_score(y_test,pred_lr)
for j in range(2,10):
    lsscore=cross_val_score(lr,x,y,cv=j)
    lsc=lsscore.mean()
    print("at cv",j)
    print("cross val score",lsc*100)
    print("accuracy_score",lss*100)
    print("\n")
```

```python
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier, BaggingClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC= RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
score = cross_val_score(RFC,x,y)
print(score)
```

```python
print(score.mean())
print("diff between accuracy score and cross val",accuracy_score(y_test,predRFC))
```
